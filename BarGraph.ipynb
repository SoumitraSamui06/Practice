```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\itsso\OneDrive\Documents\studentmarks.csv")
df.head()
```

```python
subject_colors = ['blue', 'green', 'red']

for i, student in enumerate(df['Name']):
    fig, ax = plt.subplots()
    
    
    ax.bar(['Maths', 'Physics', 'Chemistry'], [df.loc[i, 'Maths'], df.loc[i, 'Physics'], df.loc[i, 'Chemistry']], color=subject_colors)

    for j, subject in enumerate(['Maths', 'Physics', 'Chemistry']):
        ax.text(j, df.loc[i, subject] + 1, f"{df.loc[i, subject]}", ha='center', va='bottom')
    
    ax.set_xlabel('Subjects')
    ax.set_title(f'{student} Result')
    
    
    plt.show()
```
